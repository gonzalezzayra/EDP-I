**Ejercicio 1**

**En Cosenos**

```python
from sympy import fourier_series, pi, plot, Piecewise, exp
from sympy.abc import x
```

```python
f = Piecewise(
    ((exp(2*x)), ( -1<= x ) & (x <= 0)),
    ((exp(2*x)), ( 0<= x ) & (x <= 1))

)
```

```python
plot(f, (x,-1,1),line_color ='red')
```

```python
s = fourier_series(f, (x, -1, 1))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(f, s1, s2, s3, (x, -1, 1), show=False, legend=True)
```

```python
s
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
```

```python
p.show()
```
