```python
#Importar librerias
import matplotlib.pyplot as plt
import numpy as np

#Graficar
x_axis = np.linspace(-3,3)
def square(x):return x**2
y_axis = square(x_axis)
plt.plot(x_axis, y_axis, color="green")
```
