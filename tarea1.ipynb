**Series de Fourier**

Graficar la función definida a trozos

$$f(x)= 0 \hspace{3pt} si \hspace{3pt} -2\leq x \leq 0 $$


$$f(x)= 2-x \hspace{3pt} si \hspace{3pt} 0\leq x \leq 2 $$

Esta dada en términos de la serie de Fourier por la expresión:
$$f(x)=\frac{1}{2} + \sum_{n=1}^{\infty}\frac{2}{{n^2}{\pi}} (1-(-1)^n) cos(\frac{n\pi x}{2} )\frac{(-1)^{n+1}}{n}  + \frac{2}{n \pi}\sin(\frac{n\pi x}{2}).$$

```python
from pylab import *
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
def recta(x):
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] = 2 - x[i]
    return y
```

```python
# Crear un arreglo de valores de x entre 0 y 10
x = linspace(-2, 2, 100)
f = recta(x)  # Llamamos a la función recta

# Graficamos el resultado
plot(x, y)
xlabel('x')
ylabel('y')
grid()
title('Gráfico de y = 2 - x')
show()
```

```python
def u_0(x,N=20): #N es el número de amornicos
  f = zeros(len(x))
  for n in range(1,N):
    f += (2/((n**2)*(pi**2)))*(1-((-1)**n))*cos((n*pi*x)/2) + (2/(n*pi))*sin((n*pi*x)/2)
  return f+(1/2)
```

f_n = u_0(x)

```python
f_n = u_0(x)
plot(x,f_n,label='$f_N(x)$ con $N=20$')
plot(x,f,label='$f(x)=2-x$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-2,1.3]),ylim([0.8,2.5]),grid(),legend()
show()
```

```python
for i in range(1,20,5):
    plot(x,u_0(x,i),label='$f_N(x)$ con $N=%d$' %i)
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-0.01,1.01]),ylim([-0.8,3]),grid(),legend()
show()
```
